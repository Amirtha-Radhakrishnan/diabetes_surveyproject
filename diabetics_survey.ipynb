```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# we should always specify the name of the document along with dataset's name

df =pd.read_csv("C:\\Users\\Lenovo\\OneDrive\\Desktop\\diabetic_datasheet\\diabetes.csv")
```

```python
## to print complete dataset

print(df)
```

```python
# similarly to print dataset in a structured manner

pd.DataFrame(df)
```

```python
# to delete a null value in the dataframe (all rows and columns) we can use dropna()
## it will alter the original dataset

df.dropna(inplace=True)
```

```python
# to get descriptive statistics

df.describe()
```

```python
pd.DataFrame(df)
```

```python
# using histplot 

data =pd.DataFrame(df)
#plt.figure(figsize=(8,6))
sns.histplot(data,x="Age",bins=10,color="Red",kde=True)
plt.show()
sns.histplot(data["Glucose"],bins=5,color="green",kde=True)
plt.show()
sns.histplot(data["BloodPressure"],bins=5,color="black",kde=True)
#plt.xlabel("Count")
#plt.ylabel("frequency")
plt.show()
#help plot
#Cell In[134], line 13
```

```python
# plotting bar graph between age and outcome
# here data = data, this refers to complete datas in the datasheet otherwise DataFrame

sns.barplot(x="Outcome",y="Age",data= data)
plt.xlabel("outcome")
plt.ylabel("Age of the women")
plt.title("outcome VS age")
plt.show()
```

```python
# for pairplot

sns.pairplot(hue='Outcome',palette ="coolwarm",data =data)
plt.show()
```

```python
plt.scatter(x="Age",y="DiabetesPedigreeFunction",c= "pink",marker='*', data =data)
plt.xlabel("Age")
plt.ylabel("DiabetesPedigreeFunction")
plt.title("Age VS DiabetesPedigreeFunction")
plt.legend()
```

```python
# heatmap

corr_matrix =data.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix,cmap="coolwarm",linewidths=.5)
plt.title("correlation heatmap")
plt.show()
```

```python
# no.of.women who is diabetic

sns.countplot(x="Outcome",palette="coolwarm",data= data)
plt.title("Outcome",fontsize= 15)
plt.show()
```

```python
# boxplot for Pregnancies VS Glucose
glucose_data = data['Glucose']
pregnancies_data = data['Pregnancies']
plt.figure(figsize=(10, 6))
plt.scatter(pregnancies_data,glucose_data)
plt.ylabel('Glucose')
plt.xlabel('Pregnancies')
plt.title('Pregnancies VS Glucose')
plt.show()
```

```python
sns.histplot(x="BMI",bins=10,data=data,kde= True)
plt.show()
```

```python
# catplot SkinThickness VS Insulin
sns.catplot(x="SkinThickness", y="Insulin", data=data, hue="Outcome", kind="violin", split=True, height=6, aspect=2)
plt.title('catplot SkinThickness VS Insulin')
plt.show()
```

```python
sns.lmplot(x="Insulin",y="SkinThickness",data=data,hue= "Outcome",palette="coolwarm")
plt.title("Insulin VS SkinThickness" )
plt.show()
```

```python
# jointplot for Pregnancies VS Outcome
data = pd.read_csv("C:\\Users\\Lenovo\\OneDrive\\Desktop\\diabetic_datasheet\\diabetes.csv")
sns.jointplot(x="Pregnancies", y="Outcome", data=data, kind="hex")
plt.figure(figsize=(12, 8))
plt.show()

                                        ### NOTE ###
# if "not defined" is shoot out we should once write import csv file code using pandas 
# i.e. data = pd.read_csv("C:\\Users\\Lenovo\\OneDrive\\Desktop\\diabetic_datasheet\\diabetes.csv")
```
